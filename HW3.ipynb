```python
import re
import torch
from torch.utils.data import Dataset, DataLoader

# ------------------------- LOAD FILES -------------------------
def load_file(path):
    with open(path, "r", encoding="utf-8") as f:
        return [line.strip() for line in f if line.strip()]

train = load_file("train.txt")
valid = load_file("valid.txt")
test  = load_file("test.txt")

# ------------------------- CLEANING ---------------------------
def cleaned(texts):
    cleaned_text = []
    words_list = []
    pattern = re.compile(r'[^A-Za-z0-9\s]')

    for text in texts:
        clean_text = pattern.sub('', text).lower()
        words = clean_text.split()
        if not words:
            continue
        words_list.extend(words)
        cleaned_text.append(words)

    return words_list, cleaned_text

# ------------------------- VOCAB ----------------------------
def Frequency_Counting(words_list, min_freq):
    from collections import Counter
    counts = Counter(words_list)

    vocab = ["<pad>", "<unk>"] + [w for w, c in counts.items() if c >= min_freq]
    word2idx = {w: i for i, w in enumerate(vocab)}
    idx2word = {i: w for w, i in word2idx.items()}

    return word2idx, idx2word

# ------------------------- DATASETS --------------------------
class Cbow_Skipgram(Dataset):
    def __init__(self, type=None, windows_size=2, cleaned_text=None, word2idx=None):
        self.windows_size = windows_size
        self.cleaned_text = cleaned_text
        self.word2idx = word2idx
        self.type = type

        if type == 'skipgram':
            self.data = self._build_skipgram_pairs()
        elif type == 'cbow':
            self.data = self._build_cbow_pairs()
        else:
            raise ValueError("type must be 'cbow' or 'skipgram'")

    def _build_skipgram_pairs(self):
        dataset = []
        unk_idx = self.word2idx["<unk>"]

        for sentence in self.cleaned_text:
            for i, center_word in enumerate(sentence):
                center_id = self.word2idx.get(center_word, unk_idx)

                for j in range(-self.windows_size, self.windows_size + 1):
                    if j == 0:
                        continue
                    k = i + j
                    if 0 <= k < len(sentence):
                        context_word = sentence[k]
                        context_id = self.word2idx.get(context_word, unk_idx)
                        dataset.append((center_id, context_id))  # correct order

        return dataset

    def _build_cbow_pairs(self):
        dataset = []
        unk_idx = self.word2idx["<unk>"]

        for sentence in self.cleaned_text:
            for i, target in enumerate(sentence):
                context_ids = []
                for j in range(-self.windows_size, self.windows_size + 1):
                    if j == 0:
                        continue
                    k = i + j
                    if 0 <= k < len(sentence):
                        context_ids.append(self.word2idx.get(sentence[k], unk_idx))

                if not context_ids:
                    continue

                target_id = self.word2idx.get(target, unk_idx)
                dataset.append((torch.tensor(context_ids, dtype=torch.long), target_id))

        return dataset

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return self.data[idx]


# ------------------------- COLLATE FUNCTIONS --------------------------
def cbow_collate_fn(batch):
    contexts, targets = zip(*batch)
    pad_id = 0  # <pad> always index 0

    padded = torch.nn.utils.rnn.pad_sequence(
        contexts, batch_first=True, padding_value=pad_id
    )
    targets = torch.tensor(targets, dtype=torch.long)
    return padded, targets

def skipgram_collate_fn(batch):
    centers = torch.tensor([x[0] for x in batch], dtype=torch.long)
    contexts = torch.tensor([x[1] for x in batch], dtype=torch.long)
    return centers, contexts


# ------------------------- BUILD CLEANED TEXT --------------------------
words_list_train, cleaned_text_train = cleaned(train)
words_list_valid, cleaned_text_valid = cleaned(valid)
words_list_test, cleaned_text_test = cleaned(test)

# build vocab ONLY from train
word2idx, idx2word = Frequency_Counting(words_list_train, min_freq=2)

# ------------------------- DATASETS --------------------------
train_dataset_cbow = Cbow_Skipgram(
    type='cbow', windows_size=2, cleaned_text=cleaned_text_train, word2idx=word2idx
)
valid_dataset_cbow = Cbow_Skipgram(
    type='cbow', windows_size=2, cleaned_text=cleaned_text_valid, word2idx=word2idx
)
test_dataset_cbow = Cbow_Skipgram(
    type='cbow', windows_size=2, cleaned_text=cleaned_text_test, word2idx=word2idx
)

train_dataset_skip = Cbow_Skipgram(
    type='skipgram', windows_size=2, cleaned_text=cleaned_text_train, word2idx=word2idx
)
valid_dataset_skip = Cbow_Skipgram(
    type='skipgram', windows_size=2, cleaned_text=cleaned_text_valid, word2idx=word2idx
)
test_dataset_skip = Cbow_Skipgram(
    type='skipgram', windows_size=2, cleaned_text=cleaned_text_test, word2idx=word2idx
)

# ------------------------- DATALOADERS --------------------------
train_loader_cbow = DataLoader(train_dataset_cbow, batch_size=256, shuffle=True, collate_fn=cbow_collate_fn)
valid_loader_cbow = DataLoader(valid_dataset_cbow, batch_size=256, collate_fn=cbow_collate_fn)
test_loader_cbow  = DataLoader(test_dataset_cbow, batch_size=256, collate_fn=cbow_collate_fn)

train_loader_skipgram = DataLoader(train_dataset_skip, batch_size=256, shuffle=True, collate_fn=skipgram_collate_fn)
valid_loader_skipgram = DataLoader(valid_dataset_skip, batch_size=256, collate_fn=skipgram_collate_fn)
test_loader_skipgram  = DataLoader(test_dataset_skip, batch_size=256, collate_fn=skipgram_collate_fn)
```

```python
import math
def nearest_neighbors(model_emb, word_idx, idx2word, topk=5):
    emb = model_emb.weight.data.cpu()
    query = emb[word_idx].unsqueeze(0)
    norms = emb.norm(dim=1, keepdim=True)
    qnorm = query.norm()
    sims = (emb @ query.t()).squeeze(1) / (norms.squeeze(1) * qnorm + 1e-9)
    vals, inds = sims.topk(min(topk + 1, len(inds:=range(len(idx2word)))))
    neighbors = []
    for v, i in zip(vals.tolist(), inds.tolist()):
        neighbors.append((idx2word[i], v))
    neighbors = [n for n in neighbors if n[0] != idx2word[word_idx]][:topk]
    return neighbors

print("\nExample neighbors using CBOW embeddings:")
for w in ["i", "deep", "python", "learning"]:
    idx = word2idx.get(w, 1)
    print(f" {w} ->", nearest_neighbors(cbow_model.emb, idx, idx2word, topk=4))

print("\nExample neighbors using Skip-gram embeddings:")
for w in ["i", "deep", "python", "learning"]:
    idx = word2idx.get(w, 1)
    print(f" {w} ->", nearest_neighbors(skip_model.emb, idx, idx2word, topk=4))
```

```python
from datasets import DatasetDict
from collections import defaultdict
from datasets import load_dataset
import random
import re

ds = load_dataset("SetFit/ag_news")

train_set = ds['train']
test_set = ds['test']

def balanced_sample(dataset, total_samples):
    labels = list(set(dataset['label']))
    per_class = total_samples // len(labels)

    grouped = defaultdict(list)
    for idx, lbl in enumerate(dataset['label']):
        grouped[lbl].append(idx)

    sampled_indices = []
    for lbl in labels:
        sampled_indices.extend(random.sample(grouped[lbl], per_class))

    return dataset.select(sampled_indices), sampled_indices

train , _= balanced_sample(train_set, 5000)
idxs = []
valid , idxs = balanced_sample(test_set, 1000)

remain_dataset = test_set.select(
    [i for i in range(len(test_set)) if i not in idxs]
)

test,_ = balanced_sample(remain_dataset, 1000)

final_ds = DatasetDict({
    "train": train,
    "valid": valid,
    "test": test
})


def preprocess_text(example):
    text = example["text"]
    text = text.lower()

    text = re.sub(r"\s+", " ", text).strip()

    example["text"] = text
    return example

cleaned_dataset = final_ds.map(preprocess_text)
print(cleaned_dataset)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import matplotlib.pyplot as plt

class MyDataset(Dataset):
    def __init__(self, embeddings, labels):
        self.embeddings = torch.tensor(embeddings, dtype=torch.float32)
        self.labels = torch.tensor(labels, dtype=torch.long)

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return self.embeddings[idx], self.labels[idx]

embeddings_train = embedded_dataset["train"]["embedding"]
labels_train = embedded_dataset["train"]["label"]

embeddings_valid = embedded_dataset["valid"]["embedding"]
labels_valid = embedded_dataset["valid"]["label"]

embeddings_test = embedded_dataset["test"]["embedding"]
labels_test = embedded_dataset["test"]["label"]

dataset_train = MyDataset(embeddings_train, labels_train)
dataset_valid = MyDataset(embeddings_valid, labels_valid)
dataset_test  = MyDataset(embeddings_test, labels_test)

dataloader_train = DataLoader(dataset_train, batch_size=32, shuffle=True)
dataloader_valid = DataLoader(dataset_valid, batch_size=32, shuffle=True)
dataloader_test = DataLoader(dataset_test, batch_size=32, shuffle=True)


class FCNet(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, p=0.5):
        super().__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.dropout = nn.Dropout(p)
        self.fc2 = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x

device = "cuda" if torch.cuda.is_available() else "cpu"

model = FCNet(input_dim=300, hidden_dim=128, output_dim=len(set(labels_train))).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3 ,weight_decay=1e-4)

epochs = 100

train_losses, val_losses = [], []
train_accs, val_accs = [], []

for epoch in range(epochs):
    # --- Train ---
    model.train()
    total_loss, correct, total = 0, 0, 0
    for X, y in dataloader_train:
        X, y = X.to(device), y.to(device)

        optimizer.zero_grad()
        outputs = model(X)
        loss = criterion(outputs, y)
        loss.backward()
        optimizer.step()

        total_loss += loss.item() * X.size(0)
        preds = outputs.argmax(dim=1)
        correct += (preds == y).sum().item()
        total += y.size(0)

    train_loss = total_loss / total
    train_acc = correct / total
    train_losses.append(train_loss)
    train_accs.append(train_acc)

    # --- Validation ---
    model.eval()
    val_loss, correct, total = 0, 0, 0
    with torch.no_grad():
        for X, y in dataloader_valid:
            X, y = X.to(device), y.to(device)
            outputs = model(X)
            loss = criterion(outputs, y)
            val_loss += loss.item() * X.size(0)
            preds = outputs.argmax(dim=1)
            correct += (preds == y).sum().item()
            total += y.size(0)

    val_loss /= total
    val_acc = correct / total
    val_losses.append(val_loss)
    val_accs.append(val_acc)

    print(f"Epoch {epoch+1:02d} | Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} | Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}")

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(train_losses, label="Train Loss")
plt.plot(val_losses, label="Val Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss Curve")
plt.legend()

plt.subplot(1,2,2)
plt.plot(train_accs, label="Train Acc")
plt.plot(val_accs, label="Val Acc")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Accuracy Curve")
plt.legend()

plt.show()
```
